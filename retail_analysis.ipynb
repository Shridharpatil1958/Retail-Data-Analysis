```python
# Data Handling
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)

# Style
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("retail_sales_dataset_15000_rows.csv")

# View first rows
df.head()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
df.dtypes
```

```python
df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month
df['day'] = df['order_date'].dt.day
```

```python
df.head()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['sales'])
plt.title("Sales Outliers")
plt.show()
```

```python
Q1 = df['sales'].quantile(0.25)
Q3 = df['sales'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['sales'] >= lower) & (df['sales'] <= upper)]
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['sales'], bins=30, kde=True)
plt.title("Sales Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['profit'], bins=30, kde=True)
plt.title("Profit Distribution")
plt.show()
```

```python
sns.countplot(x='quantity', data=df)
plt.title("Quantity Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='category', data=df)
plt.title("Product Category Distribution")
plt.show()
```

```python
sns.countplot(x='sales_channel', data=df)
plt.title("Online vs Offline Sales")
plt.show()
```

```python
sns.countplot(x='payment_method', data=df)
plt.title("Payment Methods")
plt.show()
```

```python
category_sales = df.groupby('category')['sales'].sum().sort_values(ascending=False)

category_sales.plot(kind='bar')
plt.title("Sales by Category")
plt.show()
```

```python
region_profit = df.groupby('region')['profit'].sum()

region_profit.plot(kind='bar')
plt.title("Profit by Region")
plt.show()
```

```python
channel_sales = df.groupby('sales_channel')['sales'].sum()

channel_sales.plot(kind='bar')
plt.title("Online vs Offline Sales")
plt.show()
```

```python
monthly_sales = df.groupby('month')['sales'].sum()

plt.figure(figsize=(8,5))
monthly_sales.plot(marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
year_sales = df.groupby('year')['sales'].sum()

year_sales.plot(kind='bar')
plt.title("Yearly Sales")
plt.show()
```

```python
corr = df[['sales','profit','quantity','discount']].corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
top_products = df.groupby('subcategory')['sales'].sum().sort_values(ascending=False)

top_products.head(10)
```

```python
city_sales = df.groupby('city')['sales'].sum().sort_values(ascending=False)

city_sales.plot(kind='bar', figsize=(8,5))
plt.title("Top Cities by Sales")
plt.show()
```

```python
top_customers = df.groupby('customer_id')['sales'].sum().sort_values(ascending=False)

top_customers.head(10)
```

```python
df.to_csv("cleaned_retail_sales_data.csv", index=False)
```
